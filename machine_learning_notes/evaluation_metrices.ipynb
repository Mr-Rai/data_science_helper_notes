```python
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,
    roc_auc_score, roc_curve, classification_report
)
from sklearn.metrics import (
    mean_absolute_error, mean_squared_error, r2_score
)
model_res_data = [['model', 'precision', 'recall', 'F1 Score', 'TP, FP, FN, TN']]
```

```python
# classification
X_cls, y_cls = make_classification(n_samples=1000, n_features=10, n_informative=5,
                           n_redundant=2, n_clusters_per_class=2, random_state=42)
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_cls, y_cls, random_state=42, test_size=0.25)

classifier = LogisticRegression(max_iter=500).fit(X_train_cls, y_train_cls)
y_pred_cls = classifier.predict(X_test_cls)
y_pred_prob = classifier.predict_proba(X_test_cls)[:,1]
```

```python
# regression
X_reg, y_reg = make_regression(n_samples=1000, n_features=10, n_informative=5,noise=15.0, random_state=42)
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_reg, y_reg, random_state=42, test_size=0.25)

regressor = LinearRegression().fit(X_train_reg, y_train_reg)
y_pred_reg = regressor.predict(X_test_reg)
```

```python
def get_classification_metrics(y_test, y_pred):
    return {
        "precision": precision_score(y_test, y_pred, zero_division=0, average='macro'),
        "recall": recall_score(y_test, y_pred, zero_division=0),
        "F1 Score": f1_score(y_test, y_pred, zero_division=0),
        "cm": confusion_matrix(y_test, y_pred),
        "TP, FP, FN, TN": confusion_matrix(y_test, y_pred).ravel()
    }
```

```python

def get_regression_metrics(y_test, y_pred):
    return {
        "MAE": mean_absolute_error(y_test, y_pred),
        "MSE": mean_squared_error(y_test, y_pred),
        "RMSE": np.sqrt(mean_squared_error(y_test, y_pred)),
        "R2 Score": r2_score(y_test, y_pred)
    }
```

```python
get_classification_metrics(y_test_cls, y_pred_cls)
```

```python
get_regression_metrics(y_test_reg, y_pred_reg)
```

```python
from sklearn.naive_bayes import GaussianNB

nb_clf = GaussianNB().fit(X_train_cls, y_train_cls)
nb_pred_y = nb_clf.predict(X_test_cls)
get_classification_metrics(y_test_cls, nb_pred_y)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_clf = RandomForestClassifier().fit(X_train_cls, y_train_cls)
y_pred_rf = rf_clf.predict(X_test_cls)
print(get_classification_metrics(y_test_cls, y_pred_rf))
pd.DataFrame(get_classification_metrics(
    y_test_cls, y_pred_rf)['cm'],
    columns=['predicted True', 'Predicted False'],
    index=['True Positive', 'True Negative'])
```

#### Gausian / Normal Distribution
The Gaussian distribution, also called the normal distribution, is a continuous probability distribution that forms the famous bell‑shaped curve.
It describes how data tends to cluster around a central value (the mean) with decreasing frequency as you move away from it.

**Key Characteristics**
1. Bell-Shaped Curve
   1. Highest at the mean (center)
   2. Symmetrical on both sides
2. Two Parameters Define It
   1. Mean (μ)
      1. Center of the distribution
      2. The peak of the curve
      3. Represents the average value
   2. Standard Deviation (σ)
      1. Measures how spread out the data is
      2. Small σ → narrow/tall curve
      3. Large σ → wide/flat curve

**The Empirical Rule (68–95–99.7 Rule)**
For normal data:
- 68% of values lie within 1 standard deviation (μ ± 1σ)
- 95% lie within 2 standard deviations (μ ± 2σ)
- 99.7% lie within 3 standard deviations (μ ± 3σ)

**This is extremely useful in:**
- Outlier detection
- Quality control
- Standardized test scores

#### Z Score
A Z‑score tells you how many standard deviations a value is from the mean of a dataset.
It’s a way to standardize different data so you can compare them easily.



Where:
- x = the data value
- μ = mean
- σ = standard deviation

**Interpretation**
- z = 0 : The value is exactly equal to the mean.
- z = +1 : The value is 1 standard deviation above the mean.
- z = –2 : The value is 2 standard deviations below the mean.

**Use Cases**
- To compare values from different scales
- To detect outliers
- To convert data into the standard normal distribution
- Used in statistics, machine learning, hypothesis testing, confidence intervals, etc.

**In a normal (Gaussian) distribution:**
- 68% of values fall between z = −1 and +1
- 95% fall between z = −2 and +2
- 99.7% fall between z = −3 and +3
- Values outside this range (large |z|) are often considered outliers.

````python
import numpy as np

data = [1,2,3,4,5,6,7]
std_dev = np.std(data)
mean = np.mean(data)
print("Standard Deviation:", std_dev)
print("Mean:", mean)
z_score = {x: f'{float((x - mean) / std_dev)} SD' for x in data}
z_score
````

#### Standard Normal Distribution
The standard normal distribution is a special type of normal (Gaussian) distribution.
It has:
- Mean (μ) = 0
- Standard Deviation (σ) = 1

This is just the normal distribution after converting all values to Z‑scores. This process is called **Standardization**

**Key Features**
- Bell‑shaped and perfectly symmetric
- Centered at 0
- One standard deviation on either side is ±1
- Uses Z-scores as the scale instead of raw values

**Use Cases**
- Makes probability calculations easier
- Lets us use the Z‑table
- Helps compare different datasets
- Used in hypothesis testing, confidence intervals, and statistics

````python
Standardization_data  = [float((x - mean) / std_dev) for x in data]
Standardization_mean = np.mean(Standardization_data)
Standardization_std_dev = np.std(Standardization_data)
print("Standardized Data Mean:", Standardization_mean)
print("Standardized Data Standard Deviation:", Standardization_std_dev)

z_score_normalized = {x: f'{float((x - Standardization_mean) / Standardization_std_dev)} SD' for x in Standardization_data}
z_score_normalized
````

#### Another example of Z score use case
ODI Series Data:
1. 2020
   1. Series Average Score = 250
   2. Standard Deviation = 10
   3. Team Final Score = 240
2. 2021
   1. Series Average Score = 260
   2. Standard Deviation = 12
   3. Team Final Score = 245

In which year team performed well?

````python
z_score_for_2020 = (240 - 250) / 10
z_score_for_2021 = (245 - 260) / 12

print("Z-Score for 2020:", z_score_for_2020)
print("Z-Score for 2021:", z_score_for_2021)

if z_score_for_2020 > z_score_for_2021:
    print("The year 2020 had a better performance relative to its mean.")
else:
    print("The year 2021 had a better performance relative to its mean.")
````

#### Z Table
A Z‑table (Standard Normal Table) is a chart that shows the area (probability) under the standard normal distribution curve for a given Z‑score. In simple words:
```
A Z‑table tells you how much of the data lies to the LEFT of a Z‑score on the standard normal curve.
If you pick a Z‑score like 1.0, the Z‑table tells you:
How much of the curve is to the left of 1.0 standard deviations above the mean.
```

For ex. z = 1.0 → 0.8413 implies that 84.13% of values are below z = 1.0<br>
if you want right side area then do 1 - 0.8413




Since the standard normal distribution has:
- mean 0
- standard deviation 1

thus the Z‑table works for any normal distribution once you convert values to Z‑scores.

**How to read Z table**
Z‑score has two parts:
- Row = first two digits (like 1.2)
- Column = last decimal digit (like .03)

So for z = 1.23:
- Row: 1.2
- Column: 0.03
- Table value: 0.8907

Meaning: 89.07% of the curve is to the left of 1.23

**Steps to solve any Z-Table problem**
- **Step 1:** Convert the raw values to z score (z = x−μ / μ​)
- **Step 2:** Use the Z‑table to get the probability
  1. P(Z ≤ z) --> Just read the table (Z < z means area to the left)
  2. P(Z ≥ z) = 1 − P(Z ≤ z)
  3. Area between two Z‑scores, say z1​ and z2: P(z1 ​≤ Z ≤ z2​) = P(Z ≤ z2​) − P(Z ≤ z1​)


**Example**

What percent of values are below x = 85 if μ = 70, σ = 10 ?


````python
# solving above problem
x = 85
mean = 70
std_dev = 10

z_score = (x - mean) / std_dev
print("Z score:", z_score)

# using table, P(Z ≤ 1.5) = 0.9332
print("93.32% of area lies below 85")
````

````python
'''
Q2. In India, the average IQ is 100 with a standard deviation of 15. What percentage of population
would you expect to have an IQ lower than 85
'''

average_iq = 100
std_deviation = 15

z_score_85 = (85 - average_iq) / std_deviation
print("Z score for IQ 85:", z_score_85)

# Using Z-table, P(Z ≤ -1.0) = 0.1587
print("15.87% of population is expected to have an IQ lower than 85")
````

#### Normalization
Normalization is a scaling technique that transforms data into a fixed range, usually 0 to 1
It changes the scale of the data but does NOT change the shape of the distribution.

Most Common Type is **MinMax Normalization**



example:

data = [10, 20, 30, 40, 50]<br>
Min = 10<br>
Max = 50<br>
Normalize 30:<br>
x′ = (30 - 10) / (50 - 10)<br>
x′ = 0.5

**Key Difference From Standardization**

| Standardization (Z-score)     | Normalization (Min–Max)       |
|-------------------------------|-------------------------------|
| Mean becomes 0                | Values always between 0 and 1 |
| SD becomes 1                  | Used when scale matters (e.g., neural networks, distance‑based algorithms) |
| Values can be negative or > 1 |                               |
| Used for normal distributions |                               |


**Use normalization when:**
- You need everything between 0–1
- You want to keep the shape of the data
- You work with ML algorithms sensitive to scale (KNN, K‑means, Neural Networks)
- You are plotting data together with different units

**Avoid normalization when:**
- You have outliers → because min and max are affected
- Your data is Gaussian → use standardization instead

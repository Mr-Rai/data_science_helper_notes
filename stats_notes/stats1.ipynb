### Variable measurement scales
Dataset will have these kind of data
- Normal: Caegorical or Qualitative
- Ordinal: Order of the data matters, values does not. ex- rank of students is considered, marks is not
- Interval: Order matters, value also matters and natural zero is not present ex. (10,20), (20,30)

### Histogram
say we have data of ages
ages = [10,12,14,18,26,30,35,37,40,41,42,43,50,51]

**What are bins in histogram:** contiguous intervals (ranges) that divide the x-axis (data range).
Each bin counts how many data points fall within that range, and the y-axis shows the frequency (or density)

**What is the use of PDF:** Probability Distribution Function is used for smoothing of histogram by normalizing it that represents probabilty distribution rather than raw counts. It uses Kernal Density Estimator

**Difference between Bar Graph and Hostogram:** Bar graph is for discrete data and histogram is for continuos data

```python
# Plot histogram for above
import numpy as np
import matplotlib.pyplot as plt
ages = [10,12,14,18,26,30,35,37,40,41,42,43,50,51]

'''
In matplotlib,
bins=20 does NOT mean bin width = 20
It means number of bins = 20

Matplotlib automatically looks at the minimum and maximum of your data,
divides that range into 20 equal-width bins
'''

plt.hist(ages, bins=10, edgecolor='black')
plt.show()
```

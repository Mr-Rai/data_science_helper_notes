### Measure of Central Tendency
Refers to the measure used to determine the center of the distribution of the data.
We can use mean, median or mode for this.

#### Median
In above data, if we add one more entry, 100. Then the mean jumps from 3.2 to 12.0. This will have adverse effect on the entire distribution.

Here value 100 is called **Outliers**

data = [1,1,2,2,3,3,4,5,5,6,100]

Steps:
1. Sort the data in ascending order
2. The median is the middle value (or average of the two middle values) depending on whether n is odd or even.
   1. n is odd then $\frac{(n+1)}{2}$ th element is median
   2. n is even then average of sum of $\frac{n}{2}$ th and $\frac{(n+1)}{2}$ th element is median

| n is even | n is odd |
|-----------|----------|
|  |  |

median = 11+1/2 = 6th element = 3

Let's add one more outlier to our data.
data = [1,1,2,2,3,3,4,5,5,6,100,110]

new median = (6th element + 7th element)/2 = (3+4)/2 = 3.5

**What we noticed:** median is less sensitive to outliers as compared to mean

---------------------------------------

#### Mode
Element with most no. of occurence in our data is Called Mode of data

data = [1,1,2,2,3,3,4,5,5,6,6,6,100,110]

mode is 6 with max no. of count 3

**Problem:** if we add 3 more entries for 100 then mode will be 100 although it was outliers

**Where mode is used:** Suppose we have a categorial dataset (ex. List of flowers) where some entries are missing then in that case we can replace these values with mode

```
Take another example, we have students age dataset where some entries are missing then we will use mean. It's age of given students group (~5 to 25) so it cannot have very big outlier that could affect overall mean
```




````python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = np.array([2, 4, 5, 7, 9, 10])

# Calculate statistics
mean = np.mean(data)
variance = np.var(data)
std_dev = np.std(data)

# Plotting the data points
plt.figure(figsize=(10, 4))
plt.scatter(range(len(data)), data, color='blue', label='Data points', zorder=5)

# Plot mean line
plt.axhline(mean, color='red', linestyle='--', label=f'Mean (μ) = {mean}')

# Show standard deviation lines
plt.axhline(mean + std_dev, color='green', linestyle='-.', label=f'Mean + SD = {mean + std_dev}')
plt.axhline(mean - std_dev, color='green', linestyle='-.', label=f'Mean - SD = {mean - std_dev}')

# Annotate variance visually (squared distances can be illustrated)
for i, y in enumerate(data):
    plt.plot([i, i], [mean, y], color='purple', alpha=0.5)

plt.title('Mean, Standard Deviation, and Variance Visualization')
plt.xlabel('Data Index')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
````

### Measure of Dispersion
Dispersion (or variability) measures how spread out or scattered the values in a dataset are. It tells us whether the data points are close to the mean or widely spread.
Common Measures of Dispersion:
- Range (Range=Max−Min)
- Variance (average of the squared differences from the mean)
- Standard Deviation (square root of variance, gives spread in same units as the data)
- Mean Deviation – average of absolute differences from the mean

Formulas:

| Measure          | Formula                                |
|------------------|----------------------------------------|
|Range             |      |
|Variance          |  |
|Standard Deviation|  |
|Mean Deviation    |  |

```
Dispersion tells us whether the data is tightly packed or widely spread around the average.
```

**Why do we need dispersion**: Consider two datasets
data_a = [1,1,2,2,4]
data_b = [2,2,2,2,2]

mean for both of them will be 2.0, so how we will distinguish difference between data A and data B. This is where dispersion comes into picture

**Key Points:**
- Standard deviation (SD) measures the amount of variability or dispersion in a dataset—how far values typically deviate from the mean.
- A low standard deviation indicates data points are tightly clustered around the mean, while a high standard deviation indicates greater spread.
- It is the square root of variance, which keeps the unit of measurement consistent with the original data.
- In data science, SD is used to understand data distribution, detect anomalies, and compare variability across features.
- Standard deviation is sensitive to outliers, extreme values can significantly increase it.
- It plays a key role in normalization and scaling (e.g., z-score standardization).
- SD is foundational to many statistical models and assumptions, including normal distributions and confidence intervals.
- In machine learning, features with very different standard deviations may bias model performance if not properly scaled.
---------------------------------------------------

### 1. Variance
Variance measures how much the data values deviate from the mean on average. It’s a key measure of dispersion.

**Key Points:**
- Variance is always non-negative.
- The larger the variance, the more spread out the data.
- The units of variance are the square of the original units.
- Variance is sensitive to outliers, often more so than standard deviation due to squaring.
- In data science, variance is widely used in feature selection (e.g., removing low-variance features).
- High variance in a feature indicates greater information content but may also introduce noise.
- Variance is central to the bias–variance tradeoff in machine learning models.
- Many algorithms (e.g., PCA) rely on variance to identify directions of maximum information.
- Variance is a building block for other metrics, including standard deviation, covariance, and correlation.
- In probabilistic models, variance represents uncertainty or risk associated with a random variable.

````python
import numpy as np
import matplotlib.pyplot as plt

# Sample data
data = [2, 4, 5, 7, 9, 10]

# Mean
mean = np.mean(data)
squared_devs = [(x - mean)**2 for x in data]  # squared deviations

# Plot squared deviations as bars
plt.bar(range(len(data)), squared_devs, color='purple', alpha=0.6, label='Squared deviations')

# Plot mean line
plt.axhline(np.mean(squared_devs), color='red', linestyle='--', label=f'Variance = {np.mean(squared_devs):.2f}')

plt.title('Variance Visualization (Squared Deviations)')
plt.xlabel('Data Index')
plt.ylabel('Squared Deviation')
plt.legend()
plt.show()
````

#### Percentiles and Quartiles
Used to find outliers

**Percentile:** is a value below which a certain percentage of observation lie

> percentile rank of x = (# of values below x / n)*100

where:<br>
- x is value for which we need to find percentile
- n is total no. of records in dataset

Let's understand it with below calculations.


````python
import numpy as np
data = [2,2,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,11,12]
x = 10
# percentile = np.percentile(data, 25)
# print(percentile)
percentile = len([value for value in data if value < x])/ len(data) * 100
percentile
````

So, it says 80% of values in dataset are less than 10

Another example: What value exists at percentile ranking of 25%

> value = (percentile/100) * (n+1)

Below example explains it well

````python
import numpy as np

def calculate_percentile(data, number):
    # Step 1: Sort the data
    data = sorted(data)
    val = number/100*(len(data) + 1)
    print(f'Index position: {val}')

    # It returned index position as 5.25 but we do not have 5.25th index
    # So we take average of 5th index value and 6th index value. i.e 5 and 5 (index starts from 1)

    percentile = (data[int(np.floor(val))-1] + data[int(np.ceil(val))-1])/2
    print(f'Percentile Calculated through formula: {percentile}')

    # Numpy provides built-in function to calculate percentile
    percentile_np = np.percentile(data, number)
    print(f'Percentile Calculated through Numpy: {percentile_np}')
    print('---'*20)


data = [2,2,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,11,12]
calculate_percentile(data, 25)
calculate_percentile(data, 75)
````

### Five Number Summary
- Minimum
- First Quartile (Q1)
- Median
- Third Quartile (Q3)
- Maximum

Five number summary helps us remove outliers

Steps:
- Find Q1 and Q3
- Calculate IQR (Inter Quartile Range)
- IQR = Q3 ​− Q1​
- Lower Bound = Q1​ − 1.5(IQR)
- Upper Bound = Q3 ​+ 1.5(IQR)

Where:
- Q1 is the 25th percentile of a dataset. 25% of the data values lie below Q1, and 75% lie above it.
- Q3 is the 75th percentile of a dataset 75% of the data values lie below Q3, and 25% lie above it.

How to find Q1 and Q3 mathematically:
- Arrange the data in ascending order
- Find Q2 (median)
- Q1 = median of the lower half
- Q3 = median of the upper half

Or we can use:
- np.percentile(data_array, 25)
- np.percentile(data_array, 75)

````python
import numpy as np

data = [2,2,-13,3,3,4,5,5,5,6,7,8,8,8,8,8,9,9,10,11,11,12,16,25, 30]

# Find IQR
first_quartile = np.percentile(data, 25)
second_quartile = np.percentile(data, 75)
iqr = second_quartile - first_quartile

# Find low and high outlier thresholds
lower_fence = first_quartile - 1.5 * iqr
upper_fence = second_quartile + 1.5 * iqr

# Find outliers
outliers = np.where((np.array(data) < lower_fence) | (np.array(data) > upper_fence))
outliers = [x for x in data if x < lower_fence or x > upper_fence]
print(f'First Quartile (Q1): {first_quartile} | Second Quartile (Q3): {second_quartile} | IQR: {iqr}')
print(f'Outlier Thresholds: Low = {lower_fence}, High = {upper_fence}')
print(f'Outliers: {outliers}')
print('---'*20)

# Now let's find five number summary after removing outliers
cleaned_data = [x for x in data if x >= lower_fence and x <= upper_fence]
five_number_summary = {
    'Minimum': np.min(cleaned_data),
    'Q1': np.percentile(cleaned_data, 25),
    'Median': np.median(cleaned_data),
    'Q3': np.percentile(cleaned_data, 75),
    'Maximum': np.max(cleaned_data)
}
print('Five Number Summary after removing outliers:')
for key, value in five_number_summary.items():
    print(f'{key}: {value}')
````

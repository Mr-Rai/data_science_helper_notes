#### Type 1 and Type 2 error

| Null Hypothesis H₀ is | In Reality H₀ is | Decision Type | Error Type |
|-----------------------|------------------|---------------|------------|
| Rejected              | False            | Good          | NA         |
| Rejected              | True             | Bad           | Type 1     |
| Accepted              | False            | Bad           | Type 2     |
| Accepted              | True             | Good          | NA         |


It's used in **Confusion Matrix**

|          | Predicted P | Predicted N |
|----------|-------------|-------------|
| Actual T |TP           |TN           |
| Actual F |FP           |FN           |

Where:
- T/F: True, False
- P/N: Positive, Negative

Here:
- FP: In reality H₀ was True but it's rejected (Type 1)
- TN: In reality H₀ was False but it's Accepted (Type 2)

### 1 tail and 2 tail 

1. One tailed: A one‑tailed test checks for an effect in one specific direction only. We use it when we want to check if a value is greater OR less, not both.
   1. Right-Tailed test: test if the mean is greater than a value.
      1. Hypotheses: H0​: μ ≤ μ0 and H1​: μ > μ0​​
      2. p‑value formula: p = P(Z ≥ z)
    2. Left-Tailed test: test if the mean is less than a value.
      1. Hypotheses: H0​: μ ≥ μ0 and H1​: μ < μ0​​
      2. p‑value formula: p = P(Z ≤ z)
2. Two tailed: A two‑tailed test checks if the value is different — could be greater or less. We use it when we only want to know if there is any change (up or down).
   1. Hypotheses: H0​: μ = μ0 and H1​: μ != μ0​​
   2. p‑value formula: p = 2×P(Z ≥ ∣z∣)

Where:
- μ : Population Mean (The true mean of the entire population)
- μ0 : Null Hypothesis Mean (The mean assumed under the null hypothesis H₀)
- Z : Z‑score for hypothesis testing (The standardized value used to decide whether to reject H₀)
- z :  Observed Z‑value / Calculated Z‑score (The numeric value of test statistic)




**example:** 
1. **Two Tailed:** College in Karnatake have a 85% placement rate. A new college was recently opened and it was found that a sample of 150 Students had a placement rate of 88% with a standard deviation of 4%. Does this college has a different placement rate?
2. **One Tailed:**
   1. **Right tailed:** Same question but now find if this college has a **greater** placement rate?
   2. **Left tailed:** Same question but now find if this college has a **Smaller** placement rate?


### Confidence Interval (CI)
A Confidence Interval is a range of values that is likely to contain the true population parameter (such as the mean), based on sample data. It expresses uncertainty in estimation.

----------------------

**Interpretation:** A 95% confidence interval means: If we take many samples, 95% of such intervals will contain the true population mean.

**It does NOT mean:** There is a 95% chance the true mean lies in this interval.

**What affects CI width:**
- Higher confidence level → wider CI
- Larger sample size → narrower CI
- Larger variance (σ or s) → wider CI

**Example with Z-test:** On the quant exam of CAT exam, the standard deviation is known to be 100. A sample of 25 test taker has a mean of 520 score. Construct a 95% CI about the mean.

Given:
- σ = 100
- n = 25
- α = 0.05
- x bar = 520



Zα/2 ​ = Z0.025 ​= 1.96<br>
Standard Error (SE) = n/sqrt(​σ) ​= 100/sqrt(25) ​= 20​<br>
Margin of Error (ME) = Zα/2​⋅SE = 1.96×20 = 39.2<br>
Lower bound CI = 520 − 39.2 = 480.8<br>
Upper bound CI = 520 + 39.2 = 559.2<br>

We are 95% confident that the true mean CAT quant score lies between 480.8 to 559.2

**Example with t-test:** On the quant exam of CAT exam, a sample of 25 test taker has a mean of 520 with the standard deviation of 80. Construct a 95% CI about the mean. (Note: Population Standard Deviation is not given)

Given:
- s = 80 (sample SD)
- n = 25
- α = 0.05
- x bar = 520
- Degrees of freedom: df = n−1 = 24

**Step 1:** Find the Critical t‑value<br>
tα/2,24​ = t0.025,24​<br>
Note: α/2 = 0.025, then check in one tail for two tails we need to check in 0.05 of t-table

**Step 2:** Compute Standard Error (SE)<br>
SE = ​s/sqrt(n) ​= ​80/sqrt(25​) ​=16

**Step 3:** Compute Margin of Error (ME)
ME = tα/2,24​ * SE = 2.064×16 = 33.024

Lower bound CI = 520 − 33.024 = 486.976
Upper bound CI = 520 + 33.024 = 553.024

We are 95% confident that the true mean CAT quant score (when σ is unknown) lies between: 486.976 and 553.024

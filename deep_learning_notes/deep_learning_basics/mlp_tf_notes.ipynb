## TensorFlow + Keras Basics & MLP on MNIST
1. **What is TensorFlow:** TensorFlow 2.x is an open‑source deep learning framework by Google. It supports:
   1. Eager Execution (like normal Python)
   2. GPU/TPU acceleration
   3. Automatic differentiation (tf.GradientTape)
   4. Production-level deployment (TF Serving, TF Lite, TF.js)

> Think of TensorFlow as the engine that handles tensors, gradients, and hardware optimization.

2. **What is Keras:** Keras is the high‑level API for TensorFlow. It makes building neural networks super simple. Keras provides:
   1. Layers (Dense, Conv2D, LSTM, etc.)
   2. Models (Sequential, Functional, Subclassed)
   3. Training (model.fit)
   4. Callbacks (EarlyStopping, ModelCheckpoint)
   5. Optimizers (SGD, Adam, RMSprop)
   6. Loss functions
   7. Metrics (accuracy, precision, recall)

3. **Tensor & Automatic Differentiation**
   1. **What is a Tensor:** A tensor is just a multi-dimensional array.Examples:
      1. Scalar → shape ()
      2. Vector → shape (3,)
      3. Matrix → shape (3,4)
      4. Images → shape (batch, height, width, channels)
   2. **Automatic Differentiation:** TensorFlow tracks operations on tensors and computes gradients automatically.
        ```python
        with tf.GradientTape() as tape:
            y = w * x + b      # forward pass
        grad = tape.gradient(y, [w, b])  # backward pass
        ```
4. **Keras Model Types:**

    | API                       | Functionality                                     |
    |---------------------------|---------------------------------------------------|
    | Sequential API (easiest)  | Layers stacked linearly.                          |
    | Functional API            | When you need branching/skipping (e.g., ResNet).  |
    | Model Subclassing         | Define your own forward pass. Most flexible.      |

5. **MLP (Multi-Layer Perceptron) Basics**
   - MLP = fully-connected neural network.
   - `Input (784)` → `Dense(256, relu)` → `Dense(128, relu)` → `Dense(10)`
   - ASCII view: `784` → `[Dense + ReLU]` → `[Dense + ReLU]` → `[Logits]`
   - MNIST digits (0–9) are 28×28 = 784 values.

## CNNs in TensorFlow + CIFAR‑10

### CNN Fundamentals
**What a Convolution Does:** A Conv layer slides a filter over an image to detect patterns like edges, curves, and textures.

`Input Image (32x32x3)` → `conv (3x3x3 filter)` → `Feature Map (30x30x32)`

**Key Components**
- Filters/Kernels → learnable pattern detectors
- Stride → how far the filter moves
- Padding → whether the output shrinks
- Feature Maps → outputs representing detected features
- Pooling → reduces spatial size, keeps important info

### CIFAR‑10 Dataset
- 60,000 RGB images, 10 classes
- Shape: (32, 32, 3)
- Classes: airplane, dog, truck, ship, etc.

### CNN Layers in Keras
1. **Conv2D(filters, kernel_size, activation):** Learns patterns.
2. **MaxPooling2D():** Downsamples feature maps.
3. **Dropout(rate):** Regularization.
4. **BatchNormalization():** Normalizes activations, stabilizes training.
5. **Flatten() → Dense():** MLP head for classification.

### CNN Model Architecture
`Input (32x32x3)` → `Rescaling(1/255)` → `Conv2D(32, 3) + ReLU` → `Conv2D(32, 3) + ReLU` → `MaxPool` → `Dropout`

↓

`Conv2D(64, 3) + ReLU` → `Conv2D(64, 3) + ReLU` → `MaxPool` → `Dropout`

↓ 

`Flatten` → `Dense(128, ReLU)` → `Dense(10 logits)`

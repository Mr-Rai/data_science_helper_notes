## CNN in PyTorch (CIFAR‑10)

### PyTorch‑specific differences for CNN training. Unlike TensorFlow/Keras:
1. Data Pipeline
   - Uses Dataset + DataLoader
   - Augmentation done via torchvision.transforms
   - Normalization values for CIFAR‑10 must be defined manually
   - (mean = 0.4914, 0.4822, 0.4465 / std = 0.2470, 0.2435, 0.2616)
2. CNN Layers in PyTorch
   - nn.Conv2d(in_channels, out_channels, kernel_size, padding)
   - nn.BatchNorm2d(num_features)
   - nn.MaxPool2d(kernel_size)
   - nn.Dropout(p)
   - nn.Flatten() or manual .view()

3. Model Definition
   - Create a class extending nn.Module
   - __init__ = declare layers
   - forward() = define computation steps

4. Loss / Optimizer
   - Use nn.CrossEntropyLoss() (expects logits, no softmax needed)
   - Common optimizers: Adam, SGD, RMSprop
   - Weight decay = L2 regularization (passed inside optimizer)

5. Training Loop
   - Must be written manually:
   - forward → loss → backward → optimizer.step()

6. Evaluation Mode
   - model.eval() disables dropout + uses BN running stats
   - wrap in torch.no_grad()

7. Learning Rate Scheduling
   - Scheduler must be manually stepped:
   - scheduler.step()

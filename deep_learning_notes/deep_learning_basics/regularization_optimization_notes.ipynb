## Regularization, Optimizers & Learning Rate Scheduling

### Regularization Techniques
1. **Dropout:**
    1. Randomly zeros out activations during training.
       1. Helps prevent neurons from co-adapting
       2. Does not affect inference (disabled during eval)
    2. Implementation differences:
       1. PyTorch: nn.Dropout(p)
       2. TF/Keras: layers.Dropout(rate)

2. **L2 Regularization (Weight Decay)**
   1. Penalizes large weights; improves generalization.
   2. Key difference:
      1. TensorFlow: added at layer level → kernel_regularizer=l2(...)
      2. PyTorch: added in optimizer → Adam(..., weight_decay=1e-4)

3. **Early Stopping**
   1. Stops training when validation loss stops improving.
   2. Framework difference:
      1. TensorFlow: built‑in callback
      2. PyTorch: manual implementation


### Optimizers
1. **SGD:** Basic gradient descent with optional momentum.
   - Good for large datasets
   - More stable, but slower
   - Key hyperparams:
     1. lr
     2. momentum
     3. nesterov (optional)

2. **Adam:** Adaptive learning rate per parameter.
   - Fast convergence
   - Great default choice
   - Often overfits → combine with weight decay

3. **RMSProp:** Maintains moving average of squared gradients.
   - Popular for RNNs
   - Faster on noisy gradients
   - Similar to Adam without bias correction


### Learning Rate Scheduling
Scheduling affects convergence more than architecture sometimes.

1. **Step Decay:** Reduce LR every few epochs.
   - Good for stable training
2. **Cosine Annealing:** Smoothly decreases LR; restarts optional.
3. **Reduce-on-Plateau:** Reduce LR when validation metric stops improving.

--------------

**Framework Differences**

| Scheduler         | TensorFlow                                | PyTorch           |
|-------------------|-------------------------------------------|-------------------|
| Step decay        | ExponentialDecay, PiecewiseConstantDecay  | StepLR            |
| Reduce on plateau | ReduceLROnPlateau                         | ReduceLROnPlateau |
| Cosine annealing  | CosineDecay, CosineDecayRestarts          | CosineAnnealingLR |

--------------------------

## Code Examples for TensorFlow

1. **Dropout & L2**
    ```python
    model = models.Sequential([
        layers.Conv2D(32, 3, activation='relu',
                    kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
        layers.Dropout(0.3),
        layers.Flatten(),
        layers.Dense(128, activation='relu',
                    kernel_regularizer=tf.keras.regularizers.l2(1e-4)),
        layers.Dropout(0.3),
        layers.Dense(10)
    ])
    ```
--------------------

2. **Optimizer Variants**
    ```python
    optimizer = tf.keras.optimizers.SGD(learning_rate=0.01, momentum=0.9)

    # Or Adam
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3)

    # Or RMSProp
    optimizer = tf.keras.optimizers.RMSprop(learning_rate=1e-3)
    ```
--------------------

3. **Scheduling**
    ```python
    lr_schedule = tf.keras.optimizers.schedules.ExponentialDecay(
        initial_learning_rate=1e-3,
        decay_steps=10000,
        decay_rate=0.9
    )

    optimizer = tf.keras.optimizers.Adam(learning_rate=lr_schedule)
    ```
--------------------

4. **Early Stopping**
    ```python
    callback = tf.keras.callbacks.EarlyStopping(
        patience=3,
        restore_best_weights=True
    )

    model.fit(..., callbacks=[callback])
    ```

--------------------------------

## Code Examples for PyTorch

1. **Dropout & Weight Decay**
   ```python
   optimizer = optim.Adam(
      model.parameters(),
      lr=1e-3,
      weight_decay=1e-4   # L2 regularization
   )
   ```
--------------------

2. **SGD Momentum**
   ```python
   optimizer = optim.SGD(
      model.parameters(),
      lr=0.1,
      momentum=0.9
   )
   ```
--------------------

3. **RMSProp**
   ```python
   optimizer = optim.RMSprop(
      model.parameters(),
      lr=1e-3,
      alpha=0.99,
      momentum=0.9
   )
   ```
--------------------

4. **Learning Rate Scheduling**
   1. **Step decay:**
      ```python
      scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=5, gamma=0.5)
      ```

   2. **Cosine annealing:**
      ```python
      scheduler = optim.lr_scheduler.CosineAnnealingLR(
         optimizer,
         T_max=10
      )
      ```

   3. **Reduce on plateau:**
      ```python
      scheduler = optim.lr_scheduler.ReduceLROnPlateau(
         optimizer,
         mode='min',
         factor=0.5,
         patience=2
      )
      ```

   4. **Call accordingly:**
      ```python
      scheduler.step()
      # or
      scheduler.step(validation_loss)
      ```

# Deep Learning Notes: Foundations

1. **What is Deep Learning?**
   - Deep Learning = Machine Learning using neural networks with many hidden layers.
   - Key advantages
     1. Learns features automatically (no manual feature engineering).
     2. Scales extremely well with data + compute.
     3. Supports transfer learning (fine-tuning pre-trained models).
     4. Capable of solving complex tasks (vision, NLP, speech).

2. **Basic Building Blocks**
   1. **Neuron (Perceptron):** Each neuron performs:
      - $z = w \cdot x + b$
      - $a = activation(z)$

      Where:
      - w = weights
      - b = bias
      - a = output after applying activation

  2. **Activation Functions:** Purpose: Add non-linearity so the network can learn complex patterns.

      | Activation | Use Case                             |
      |------------|--------------------------------------|
      | ReLU       |Most hidden layers in modern networks |
      | Sigmoid    | Binary classification output         |
      | Softmax    | Multi-class output                   |
      | Tanh       | Some RNN architectures               |

      **Practical Note:** Always `start with ReLU` unless there is a specific reason not to.

3. **Neural Network Architecture**<br>
A typical feed‑forward NN:
`Input` → `Dense Layer` → `Activation` → `Dense Layer` → `Activation` → `Output`

**Terminology:**
- Input Layer: raw data (images, vectors, text embeddings)
- Hidden Layers: computations, linear + non-linear transformations
- Output Layer: final prediction

4. **Forward & Backward Pass**
   1. Forward Pass: Data flows through layers → produces prediction $\hat{y}$.
   2. Loss Function: Measures how wrong the prediction is. Common choices:
      - Binary Cross Entropy → binary classification
      - Categorical Cross Entropy → multi-class
      - MSE/MAE → regression
   3. Backward Pass: Gradients are calculated using backpropagation.
   4. Optimization: Weights updated using gradient-based optimizers:

   | Optimizer | Use Case                                      |
   |-----------|-----------------------------------------------|
   | SGD       | Simple, stable                                |
   | Adam      | Most commonly used; works well out of the box |
   | RMSProp   | Often used for RNNs                           |

5. **Important Training Concepts**
   1. **Batch Size:** Number of samples per weight update. Guidelines:
      - Small batch (32–64): more noise, more generalization
      - Large batch (256–1024): faster training but risks overfitting
      - Tip: Start with 64 or 128.
   2. **Learning Rate (LR):** Most important hyperparameter.
      - Too high → divergence
      - Too low → slow training
      - Typical range: $1.e^{‑4}$ to $1.e^{‑2}$
   3. **Epoch:**
      - One full pass over the dataset.
      - Use Early Stopping to avoid overfitting.
   4. **Overfitting & Fixes**
      - Overfitting = model memorizes training data.
      - Solutions:
        1. Dropout
        2. L2 Regularization
        3. Data Augmentation (images)
        4. Early Stopping
        5. Batch Normalization

6. **Types of Deep Learning Architectures**

   | Architecture          | Best For                             | Notes                          |
   |-----------------------|--------------------------------------|--------------------------------|
   | MLP / Dense Networks  | Tabular data                         | Simple, fast                   |
   | CNN                   | Images                               | Convolutions, pooling          |
   | RNN, LSTM, GRU        | Sequential data (time series, text)  | Capture temporal dependencies  |
   | Transformers          | NLP, vision, speech                  | State-of-the-art               |
   | Autoencoders          | Compression, anomaly detection       | Learn latent representations   |
   | GANs                  | Image generation                     | Generator + discriminator      |

7. **Deep Learning Tools & Frameworks**
   1. **PyTorch (most used by developers)**
      - Dynamic computation graph
      - Clean, pythonic
      - Excellent for research and production

         Basic skeleton:
         ```python
         import torch.nn as nn
         class Model(nn.Module):    
            def __init__(self):        
               super().__init__()        
               self.fc = nn.Linear(784, 10)    
               def forward(self, x):        
                  return self.fc(x)
         ```

   2. **TensorFlow / Keras**
      - Production-oriented
      - Excellent for pipelines, deployment, mobile

8. **Practical Rules**
   1. Rule 1 - Always normalize your data
      - Images → divide by 255
      - Numerical → StandardScaler / MinMax
   2. Rule 2 - Use GPU whenever possible
      - Training on CPU is painfully slow.
   3. Rule 3 - Start with a small model.
      - Avoid wasting hours training huge networks initially.
   4. Rule 4 - Use transfer learning if data < 50k
      - Especially for images and NLP.
   5. Rule 5 - Track everything (TensorBoard, WandB)
      - Helps debug training behavior.

## Scikit-learn Pipelines
In the context of machine learning, utilizing scikit-learn pipelines is essential for creating efficient and reproducible workflows. This reading explains the core components of scikit-learn pipelines, focusing on the Pipeline and ColumnTransformer classes. By understanding these components, learners will be better equipped to construct feature engineering pipelines that effectively handle various data types.

**Core Components of Scikit-learn Pipelines**

1. **Pipeline Class:** The Pipeline class in scikit-learn allows users to chain multiple processing steps together into a single workflow. This is particularly useful for organizing the sequence of transformations and ensuring that they are applied consistently during both training and evaluation. For instance, a typical pipeline might include steps for data preprocessing, feature extraction, and model fitting, all of which are encapsulated within a single object. This approach simplifies the process of applying the same transformations to new data, enhancing reproducibility.

    ```python
    from sklearn.pipeline import Pipeline
    from sklearn.preprocessing import StandardScaler
    from sklearn.linear_model import LogisticRegression
    pipeline = Pipeline([
        ('scaler', StandardScaler()),
        ('clf', LogisticRegression(max_iter=1000))
    ])
    ```

2. **ColumnTransformer Class:** The ColumnTransformer class enables users to apply different preprocessing techniques to different columns within a dataset simultaneously. This is particularly useful when dealing with mixed data types, such as numerical and categorical features. By specifying which transformers apply to which columns, users can create a unified workflow that effectively processes diverse data types.

    ```python
    from sklearn.compose import ColumnTransformer
    from sklearn.preprocessing import OneHotEncoder, StandardScaler
    from sklearn.feature_extraction.text import TfidfVectorizer
    preprocessor = ColumnTransformer(
        transformers=[
            ('num', StandardScaler(), ['numerical_column1', 'numerical_column2']),
            ('cat', OneHotEncoder(handle_unknown='ignore'), ['categorical_column']),
            ('txt', TfidfVectorizer(max_features=5000), 'text_column')],
            remainder='drop')
    ```

   1. **Tip:** If your pipeline outputs a sparse matrix (e.g., from OneHotEncoder or TfidfVectorizer), keep StandardScaler(with_mean=False) for numeric features to avoid dense conversions. If you prefer dense numeric scaling, leave it as is if your numeric branch remains dense.

3. **Chaining Steps:** The ability to chain steps in a pipeline facilitates a streamlined approach to building machine learning workflows. Each step in the pipeline can be a transformation or a model, and they are executed in the order specified. This enhances clarity while reducing the chances of errors that can occur when manually applying transformations.

4. **Applying Different Transformers:** By utilizing the ColumnTransformer, users can apply different transformers to different data columns within a single process. For example, numerical columns can undergo scaling (StandardScaler) while categorical columns are one-hot encoded (OneHotEncoder(handle_unknown="ignore")), and text columns can undergo vectorization (TfidfVectorizer); all within the same pipeline. This capability is crucial for preparing data for machine learning models that require specific input formats.

5. **Data-Leakage Guardrails**
   1. Split before fitting: Do not fit transformers on the full dataset.
   2. train_test_split → pipeline.fit(X_train, y_train) → evaluate on X_test.
   3. Encapsulate preprocessing within the pipeline: No ad-hoc transforms outside the pipeline.
   4. Use stratified splits for classification when classes are imbalanced.

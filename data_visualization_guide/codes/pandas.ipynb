```python
import pandas as pd

df = pd.read_csv('../Using Pandas/Two+Machines.csv')
df.head()
```

```python
from pydataset import data
df = data('Housing')
df.head()
```

```python
df.plot()
```

```python
df.plot(
    y='price',
    title='House Prices',
    xlabel='House Index',
    ylabel='price',
    legend=False,
    figsize=(15,4))
```

```python
df.plot(
    y=['price', 'lotsize'],
    title='House Prices',
    xlabel='House Index',
    ylabel='price',
    legend=True,
    figsize=(10,3))
```

```python
df.plot(
    y=['price', 'lotsize'],
    title='House Prices and Lot Size',
    xlabel='House Index',
    ylabel='price',
    legend=True,
    subplots=True,
    layout=(1,2),
    figsize=(10,4),
    grid=True)
```

```python
df.plot(
    y=['price', 'lotsize'],
    title='House Prices',
    xlabel='House Index',
    ylabel='price',
    legend=True,
    subplots=True,
    layout=(1,2),
    figsize=(10,3))
```

```python
# let's remove overlapping using plt.tight_layout()
import matplotlib.pyplot as plt

df.plot(
    y=['price', 'lotsize'],
    title='House Prices',
    xlabel='House Index',
    ylabel='price',
    legend=True,
    subplots=True,
    layout=(1,2),
    figsize=(10,3))
plt.tight_layout()
```

```python

df.plot(
    y=['price', 'lotsize'],
    title='House Prices',
    xlabel='House Index',
    ylabel='price',
    legend=True,
    subplots=True,
    layout=(1,2),
    figsize=(10,3),
    sharey=True)
plt.tight_layout()
```

```python
df_bathroom_price = df.groupby('bedrooms')['price'].count()
df_bathroom_price = df.groupby('bedrooms')['price'].agg('sum')
df_bathroom_price.plot(kind='bar', figsize=(5,3))
```

```python
df.plot(
    y='price',
    kind='box',
    figsize=(5,4)
)
```

```python
df.plot(
    y='price',
    kind='hist',
    bins=50
)
```

```python
df.plot(
    y='price',
    kind='kde',
    xlim=(0,250000) # because KDE gives negative as well
)
```

```python
df.plot(
    y='price',
    x='lotsize',
    kind='scatter'
)
```

```python
stories_val_cnt = df.groupby('stories')['price'].count()
stories_val_cnt = df['stories'].value_counts()
stories_val_cnt
```

```python
stories_val_cnt.plot(title='Price wrt stories', kind='pie', legend=True)
```

| Variable Types    | Plots                     |
|-------------------|---------------------------|
| Time Series       | Line Plot                 |
| One Discrete      | Bar, Pie                  |
| One Continuous    | Hist, KDE, Box, Whisker   |
| Two Continuous    | Scatter Plot              |
| One Continuous, One Discrete  | Box and Whisker (Multiple) |

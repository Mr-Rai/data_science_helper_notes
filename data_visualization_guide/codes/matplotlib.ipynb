```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pydataset import data
```

```python
df = data('austres')
df.head()
```

## Line Plot

```python
df.plot(
    x='time',
    y='austres',
    kind='kde'
)
```

```python
plt.figure(figsize=(8,3))
plt.plot(df['time'], df['austres'], 'o-', color='red')
plt.title('australian resident population')
plt.xlabel('year')
plt.ylabel('australion population')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(df['time'], df['austres'], '--b', label='population')
plt.title('australian resident population')
plt.xlabel('year')
plt.ylabel('australion population')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(df['time'], df['austres'], '-b', label='population')
plt.plot(df['time'], df['austres']+1000, '-g', label='population + 1000')
plt.title('australian resident population')
plt.xlabel('year')
plt.ylabel('australion population')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,3))
plt.plot(df['time'], df['austres'], '-b', label='population')
plt.plot(df['time'], df['austres']+1000, '-g', label='population + 1000')
plt.title('australian resident population')
plt.xlabel('year')
plt.ylabel('australion population')
plt.legend()
plt.grid(axis='y', alpha=0.5, ls='--')
plt.tight_layout()
plt.show()
```

## Bar Plot

```python
df = data('Housing')
df.head()
```

```python
bedroom_count = df.groupby('bedrooms')['bedrooms'].count()
bedroom_count
```

```python
plt.figure(figsize=(6,3))
plt.bar(
    x=bedroom_count.index,
    height=bedroom_count,
    label='bedrooms')
plt.title("Houses by number of bedrooms")
plt.xlabel("Number of bedrooms")
plt.ylabel("Counts")
plt.grid(axis='y', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
bedroom_count = df.groupby('bedrooms')['bedrooms'].count().sort_values(ascending=False)
bedroom_count
```

```python
plt.figure(figsize=(6,3))
plt.bar(
    x=bedroom_count.index.astype(str),
    height=bedroom_count,
    label='bedrooms'
)
plt.title("Houses by number of bedrooms")
plt.xlabel("Number of bedrooms")
plt.ylabel("Counts")
plt.grid(axis='y', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(
    x=bedroom_count.index.astype(str),
    height=bedroom_count,
    label='bedrooms'
)
plt.plot(
    bedroom_count.index.astype(str),
    np.cumsum(bedroom_count),
    'o-r',
    label='cummulative sum',
    alpha=0.5
)
plt.title("Houses by number of bedrooms")
plt.xlabel("Number of bedrooms")
plt.ylabel("Counts")
plt.grid(axis='y', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.bar(
    x=bedroom_count.index.astype(str),
    height=bedroom_count,
    label='bedrooms'
)
plt.plot(
    bedroom_count.index.astype(str),
    np.cumsum(bedroom_count),
    'o-r',
    label='cummulative sum',
    alpha=0.5
)
plt.axhline(0.8 * bedroom_count.sum(), alpha=0.5, color='green')
plt.title("Houses by number of bedrooms")
plt.xlabel("Number of bedrooms")
plt.ylabel("Counts")
plt.grid(axis='y', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.barh(
    y=bedroom_count.index.astype(str),
    width=bedroom_count,
    label='bedrooms'
)
plt.title("Houses by number of bedrooms")
plt.ylabel("Number of bedrooms")
plt.xlabel("Counts")
plt.grid(axis='x', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```

## Box Plot

```python
plt.figure(figsize=(8,3))
plt.boxplot(
    x=[df['price']],
    showmeans=True,
    notch=True,
    vert=False)
plt.legend()
plt.grid(alpha=0.5)
plt.title("house price distribution")
plt.show()
```

## Histogram

```python
plt.hist(
    x=df['price'],
    bins=20,
    alpha=1,
    label='price',
    histtype='bar'
)
plt.legend()
plt.xlabel('price')
plt.ylabel('counts')
plt.show()
```

```python
plt.hist(
    x=df['price'],
    bins=20,
    alpha=1,
    label='price',
    histtype='step',
    cumulative=True
)
plt.legend()
plt.xlabel('price')
plt.ylabel('counts')
plt.show()
```

## Scatter Plot

```python
plt.figure(figsize=(5,5))
plt.scatter(
    y=df['price'],
    x=df['lotsize']
)
plt.ylabel('price')
plt.xlabel('lotsize')
plt.legend()
plt.show()
```

## Pie Plot

```python
plt.pie(
    x=df['stories'].value_counts()
)
plt.legend()
plt.show()
```

```python
plt.pie(
    x=df['stories'].value_counts(),
    labels=df['stories'].value_counts().index
)
plt.legend()
plt.tight_layout()
plt.show()
```

## Categories

| Plot Type   | Meaning       | Types of Sub Plot Available                   |
|-------------|---------------|-----------------------------------------------|
| `relplot`   | relational    | scatterplot, lineplot                         |
| `displot`   | distributions | hist, kde, ecd, rug plots                     |
| `catplot`   | categorical   | strip, swarm, box, violin, point, bar plots   |

# relplot

```python
sns.relplot(
    data=df,
    x='lotsize',
    y='price',
    hue='stories',
    kind='scatter',
    aspect=2,
    height=3)
```

```python
plt.figure()
sns.relplot(
    data=df,
    x='lotsize',
    y='price',
    hue='stories',
    kind='scatter',
    aspect=2,
    height=3)
plt.title("House Price Vs Lot Size")
plt.xlabel("Lot Size of House")
plt.ylabel("Price of House")
plt.tight_layout()
plt.axhline(y=df['price'].mean())
plt.annotate(
    text="Mean Value",
    xy=(df['price'].mean(), df['lotsize'].mean())
    )
```

## cat plot
- **strip plot:** every point is plotted on graph
- **swarm plot:** every point is plotted on graph
- **box plot:** distribution level summary only like median, Q3, Q4, outliers etc
- **violin plot:** distribution level summary only like median, Q3, Q4, outliers etc
- **point plot:** just category estimates like mean, count etc
- **bar plot:** just category estimates like mean, count etc

### Strip Plot

```python
sns.catplot(
    data=df,
    x='bedrooms',
    y='price',
    kind='strip', # default
    hue='driveway',
    dodge=True,
    height=4, aspect=2
)
```

### Swarm Plot

```python
sns.catplot(
    data=df,
    x='bedrooms',
    y='price',
    kind='swarm',
    hue='driveway',
    height=5, aspect=3
)
```

### Box Plot

```python
sns.catplot(
    data=df,
    kind='box',
    y='price'
)
```

```python
sns.catplot(
    data=df,
    kind='box',
    y='price',
    x='bedrooms',
    height=4, aspect=2
)
```

### Violin Plot

```python
sns.catplot(
    data=df,
    x='bedrooms',
    y='price',
    kind='violin',
    # hue='airco'
)
```

### Bar Plot

```python
sns.catplot(
    data=df,
    x='bedrooms',
    y='price',
    kind='bar',
    # hue='driveway'
    height=3, aspect=2
)
```

```python
sns.catplot(
    data=df,
    x='bedrooms',
    y='price',
    kind='bar',
    hue='driveway',
    height=3, aspect=2,
    ci=False #Make Confidence Interval False
)
```

```python
# Count Plot

sns.catplot(
    data=df,
    x='bedrooms',
    kind='count',
    hue='driveway',
    height=3, aspect=2,
)
```

### Point Plot

```python
sns.catplot(
    data=df,
    x='bedrooms',
    y='price',
    kind='point',
    hue='driveway',
    height=3, aspect=2,
    ci=False #Make Confidence Interval False
)
```

## Joint Plots
(Scatter + Histogram)

```python
iris_df: pd.DataFrame = data('iris')
iris_df.head()
```

```python
# Scatter
sns.relplot(
    data=iris_df,
    x='Sepal.Length',
    y='Petal.Length',
    hue='Species',
    height=3, aspect=1.5
)
```

```python
# hist
sns.displot(
    data=iris_df,
    x='Sepal.Length',
    hue='Species',
    kind='kde',
    height=3, aspect=1.5
)
```

```python
sns.jointplot(
    data=iris_df,
    x='Sepal.Length',
    y='Sepal.Width',
    hue='Species',
    height=4,
)
```

## Pairplots

```python
sns.pairplot(data=iris_df)
```

```python
sns.pairplot(data=iris_df, hue='Species')
```

```python
sns.pairplot(data=iris_df, hue='Species', corner=True)
```

```python
sns.pairplot(data=iris_df,
             hue='Species',
             vars=['Sepal.Length', 'Petal.Length'])
```

## Regression or Linear Model Plot

```python
sns.lmplot(
    data=iris_df,
    x='Sepal.Length',
    y='Petal.Length',
    height=4, aspect=1.5
)
```

```python
sns.lmplot(
    data=iris_df,
    x='Sepal.Length',
    y='Sepal.Width',
    height=4, aspect=1.5
)
```

```python
sns.jointplot(
    data=iris_df,
    x='Sepal.Length',
    y='Petal.Length',
    kind='reg',
    height=4
)
```

```python
sns.pairplot(data=iris_df, kind='reg')
```

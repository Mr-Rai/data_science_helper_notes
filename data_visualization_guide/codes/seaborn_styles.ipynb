```python
import pandas as pd
import seaborn as sns
from pydataset import data
```

```python
df: pd.DataFrame = data("iris")
df.head()
```

### Without Style

```python
sns.relplot(
    data=df,
    x=df.columns[2],
    y=df.columns[3],
    hue=df.columns[-1],
    height=4, aspect=1.8
)
```

### With Style

```python
sns.set_style('whitegrid')
sns.relplot(
    data=df,
    x=df.columns[2],
    y=df.columns[3],
    hue=df.columns[-1],
    height=4, aspect=1.8
)
```

```python
sns.displot(
    data=df,
    x=df.columns[2],
    height=3, aspect=2
)
```

```python
sns.set_style('whitegrid', {'yticks.left': True})
```

```python
sns.displot(
    data=df,
    x=df.columns[2],
    height=2, aspect=1.5
)
```

```python
sns.axes_style('white')
```

# Context

```python
sns.set_context('notebook')
sns.set_style('darkgrid')
sns.displot(
    data=df,
    x=df.columns[2],
    height=4, aspect=2
)
```

# Color Palette

```python
housing_df = data("Housing")
```

```python
sns.relplot(
    data=housing_df,
    x='lotsize',
    y='price',
    hue='bedrooms',
    height=3, aspect=1.5,
    palette='bright'
)
```
